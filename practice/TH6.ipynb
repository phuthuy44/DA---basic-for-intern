```python
#ử dụng phương pháp mã hóa hàng đầu để giới hạn các giá trị tối đa và loại bỏ các giá trị ngoại lệ khỏi các biến trên.
def max_value(df3, variable, top):
    return np.where(df3[variable]>top, top, df3[variable])
columns = ['Unit Price(Currency)',
 'Exchange rate',
 'Export/Import Declaration Number',
 'HS Code',
 'Importer Code',
 'Total Value(Currency)',
 'Import Tax',
 'FOB(USD)',
 'CIF(USD)']
columns_1 =[5.499,
 24150.0,
 105451059640.0,
 63079120.0,
 13500402018.0,
 3211.8199999999997,
 0.036,
 1263.4900000000002,
 1263.4900000000002]
# Lặp qua từng cột và giá trị tối đa tương ứng
for col, max_val in zip(columns, columns_1):
    df[col] = max_value(df, col, max_val)
```
